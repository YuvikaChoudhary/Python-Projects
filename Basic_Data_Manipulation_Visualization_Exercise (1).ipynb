### 78. Create a barchart with the 10 countries that have the most 'Quantity' ordered except UK

```python
filter_countries = online_rt[online_rt['Country'] != 'United kingdom']

countries = filter_countries.groupby('Country')['Quantity'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
countries.plot(kind='bar', color='skyblue')
plt.xlabel('Country')
plt.ylabel('Total Quantity Ordered')
plt.title('Top 10 Countries by Quantity Ordered (Excluding UK)')
plt.show()
```

### 79.  Exclude negative Quatity entries

```python
positive_data = online_rt[online_rt['Quantity'] > 0]

positive_data = positive_data[positive_data['Country'] != 'United Kingdom']

top_countries = positive_data.groupby('Country')['Quantity'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
top_countries.plot(kind='bar', color='skyblue')
plt.xlabel('Country')
plt.ylabel('Total Quantity Ordered')
plt.title('Top 10 Countries by Quantity Ordered (Excluding UK and Negative Quantities)')
plt.show()
```

### 80. Create a scatterplot with the Quantity per UnitPrice by CustomerID for the top 3 Countries
Hint: First we need to find top-3 countries based on revenue, then create scater plot between Quantity and Unitprice for each country separately


```python
online_rt['Revenue'] = online_rt['Quantity'] * online_rt['UnitPrice']
country_revenue = online_rt.groupby('Country')['Revenue'].sum()
top_countries = country_revenue.sort_values(ascending=False).head(3).index
plt.figure(figsize=(12, 8))

for country in top_countries:
    country_data = online_rt[online_rt['Country'] == country]
    grouped_data = country_data.groupby(['CustomerID']).agg({'Quantity': 'sum', 'UnitPrice': 'mean'}).reset_index()
    plt.scatter(grouped_data['Quantity'], grouped_data['UnitPrice'], label=country, alpha=0.7)

plt.xlabel('Quantity')
plt.ylabel('UnitPrice')
plt.title('Quantity per UnitPrice by CustomerID for Top 3 Countries based on Revenue')
plt.legend()
plt.show()
```
